```python
import re
frequency = {}
document_text = open('turtle.txt', 'r')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
    
frequency_list = frequency.keys()
for words in frequency_list:
    print(words, frequency[words])
```

```python
import csv
from collections import Counter
import string

def calculate_word_frequency_from_csv(file_path, column_name):
    with open(file_path, 'r' , encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)

        words = []
        for row in reader:
            words.extend(row[column_name].lower().translate(str.maketrans('', '', string.punctuation)).split())

    word_frequency = Counter(words)

    for word, count in word_frequency.items():
        print(f"{word}: {count}")

file_path = 'linkedin_canada.csv'
column_name = 'location'
calculate_word_frequency_from_csv(file_path, column_name)
```
